```python
import torch

class Test:
    def __init__(self):
        self.hidden = None

    def _helper(self):
        self.hidden = torch.randn(2, 100)

    def forward(self):
        self._helper()
        hidden = self.hidden
        self.hidden = None
        return hidden
```

```python
t = Test()
t.forward()
```
